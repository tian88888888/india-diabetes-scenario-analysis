```python
import pandas as pd
from pathlib import Path

Path("../data/processed").mkdir(parents=True, exist_ok=True)
```

```python
vdem = pd.read_csv("../data/raw/vdem_liberal_democracy_index.csv")

vdem.head()
```

```python
india_vdem = vdem[vdem["Entity"] == "India"].copy()

india_vdem.head()
```

```python
india_vdem = india_vdem[["Year", "Liberal democracy index"]].rename(
    columns={
        "Year": "year",
        "Liberal democracy index": "liberal_democracy_index"
    }
)

india_vdem.head()
```

```python
india_vdem.to_csv(
    "../data/processed/india_vdem_freedom.csv",
    index=False
)
```
