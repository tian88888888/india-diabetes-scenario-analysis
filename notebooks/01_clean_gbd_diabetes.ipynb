```python
import pandas as pd
from pathlib import Path

Path("../data/processed").mkdir(parents=True, exist_ok=True)

gbd = pd.read_csv("../data/raw/ihme_gbd_india_diabetes.csv")

gbd.head()
gbd.columns
```

```python
gbd_clean = gbd[
    (gbd["location_name"] == "India") &
    (gbd["sex_name"] == "Both") &
    (gbd["age_name"] == "Age-standardized") &
    (gbd["cause_name"].str.contains("Diabetes", case=False, na=False)) &
    (gbd["metric_name"] == "Rate")
].copy()

gbd_clean = gbd_clean[
    ["year", "measure_name", "val", "upper", "lower"]
]

gbd_clean.head()
```

```python
diabetes_burden = gbd_clean.pivot_table(
    index="year",
    columns="measure_name",
    values="val"
).reset_index()

diabetes_burden.columns.name = None

diabetes_burden.head()
```

```python
diabetes_burden.to_csv(
    "../data/processed/india_diabetes_burden_clean.csv",
    index=False
)
```
